```python
from sklearn.model_selection import KFold
import pandas as pd
```

```python
covid = pd.read_csv('test_covid.csv')
covid
```

```python
#Datos covid
df_x = pd.DataFrame(covid.iloc[:,5])
df_x
```

```python
#Total Cases
df_y = covid.iloc[:,2]
df_y
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=2) #mas nsplits mas se dividen los datos

for x_values, y_values in kf.split(df_x):
    print(x_values,y_values)
    #print("Training: ",covid.iloc[x_values],"Test: ",covid.iloc[y_values])
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
regr = linear_model.LinearRegression() #Regresion lineal

#Se llama modelo de regresion lineal
```

```python
for train_index, test_index in kf.split(df_x):
    #print(train_index,test_index)
    #x_train, x_test = df_x[train_index],df_x[test_index]
    #y_train, y_test = df_y[train_index],df_y[test_index]
    #x_train = x_train.values.reshape([x_train.values.shape[0],1])
    x_main,x_test,y_main,y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=0) 
    regr.fit(x_main,y_main)
    y_pred = regr.predict(x_test)
    print(y_pred)
    print("----------------------------------------------------------------")
    r_square = metrics.r2_score(y_test, y_pred)
    print('R-Square Error associated with Random Forest Regression is:', r_square)
```
